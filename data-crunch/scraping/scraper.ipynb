```python
def get_links(url):
    html = requests.get(url).content
    soup = bs(html, "html.parser")
    a_tag = soup.find(id = "resultsCol").find_all("a")
    page_urls = [link.get("href") for link in a_tag]
    page_urls = [link for link in page_urls if "/rc" in str(link)]
    ids = []
    for link in page_urls:
        start = link.find("jk=")+3 #perch 'link.find("jk=")' mi restituisce 8, noi dobbiamo prendere la posizione 11 ---> quindi fai 8+3
        end = link.find("&fccid") #anche qui abbiamo un altro indice
        ids.append(link[start:end]) #qui prendo il link partendo da 'start' e arrivando ad 'end'
    return ids
```
