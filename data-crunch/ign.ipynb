```python
platforms = ["Xbox One", "PlayStation 4", "PC"]
platforms = rev[rev.platform.isin(platforms)]

col = ["black", "green", "orange"]

for i, plat in enumerate(platforms.platform.unique()):
    plt.figure()
    plt.hist(platforms[platforms.platform == plat].score, color = col[i])
    plt.title(plat)
    plt.xlabel("Score")
    plt.ylabel("Frequency")
```

```python
def to_action(genre):
    if "Action" in str(genre):
        return "Action"
    else:
        return genre
    
rev.genre.apply(to_action)
```

```python
rev.genre.apply(to_action).value_counts()
```

```python
def to_macro(genre):
    if "," in str(genre):
        return genre.split(",")[1].strip()
    else:
        return genre
        
rev.genre.apply(to_macro).value_counts()
```

```python
"RPG, Sport".split(",")[1].strip()
```
